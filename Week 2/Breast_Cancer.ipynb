```python
#Loading libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #for plotting graph
import seaborn as sns #for plotting graph
```

```python
column_names = ['Class','Age','Menopause','Tumor-Size','Inv-Nodes','Node-Caps','Deg-Malig','Breast','Breast-Quad','Irradiat']
df=pd.read_csv('/content/drive/MyDrive/DBMS/Breast cancer /breast-cancer.data',names=column_names)
df
```

```python
for col in df.select_dtypes(include=['object']).columns:
    unique_vals = df[col].dropna().unique()
    print(f"Column '{col}' ({len(unique_vals)} unique values):")
    print(unique_vals)
    print("-" * 50)
```

```python
for col in df.columns:
    q_count = df[col][df[col] == '?'].count()
    if q_count > 0:
        print(f"Column '{col}': '?' = {q_count}")
```

```python
import numpy as np
df_missing = df.replace("?", np.nan)
missing_proportions = df_missing.isna().mean()
for col, prop in missing_proportions.items():
    if prop:
        print(f"Column '{col}': {prop:.2%} missing (including '?')")
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype == object:
         print(col,df[col][df[col] == '?'].count())
```

```python
sns.countplot(x = "Class", data = df)
plt.title("Distribution of Class")
plt.show()

# Pie chart
df.Class.value_counts().plot.pie(autopct = "%.1f%%")
plt.title("Proportion of Class")
plt.show()
```

```python
sns.countplot(x = "Deg-Malig", data = df)
plt.title("Distribution of Deg-Malig")
plt.show()
```

```python
sns.countplot(x="Age", data = df)
plt.xticks(rotation=90)
plt.show()

print("Proportion of Age")
print(df.Age.value_counts(normalize = True)*100)
```

```python
df['Inv-Nodes'] = df['Inv-Nodes'].replace({"6-8":"3-11", "3-5":"3-11","9-11":"3-11"})
df['Inv-Nodes'] = df['Inv-Nodes'].replace({"15-17":"15-26", "12-14":"15-26","24-26":"15-26"})
```

```python
sns.countplot(x = "Inv-Nodes", data = df)
plt.title("Distribution of Inv-Nodes")
plt.show()
```

```python
df['Class']=df['Class'].replace('no-recurrence-events',0)
df['Class']=df['Class'].replace('recurrence-events',1)
df['Class'].value_counts()
```

```python
fig = plt.figure(figsize=(13,7),)
ax=sns.kdeplot(df.loc[(df['Class'] == 0),'Deg-Malig'] , color='b',shade=True,label='No rerecurrence events')
ax=sns.kdeplot(df.loc[(df['Class'] == 1),'Deg-Malig'] , color='r',shade=True, label='Recurrence events')
ax.set(xlabel='Deg-Malig', ylabel='Frequency')
plt.title('Deg-Malig VS. Class')
```

```python
#For almost every Deg-Malig, the count of 'no-recurrence-events' (blue) is significantly higher than the count of 'recurrence-events' (red).
```
